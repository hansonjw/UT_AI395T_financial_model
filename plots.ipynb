```python
import pandas as pd
from matplotlib import pyplot as plt

from data_obj import DataObj
df = DataObj().df

print(df.columns)
snp = df["Close Norm"]
```

```python
# Plot
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# rc('font',**{'family':'monospace', 'monospace':['Andale Mono']})
rc('text', usetex=True)
# 
x = snp.index
y = snp

# plot
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(x, y, linewidth=0.5, color='#00B9FF')
ax.set_xlabel("Year")
ax.set_ylabel("Normalized Price (1927=1)")
fig.suptitle("S\&P 500 Historical Price Growth")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```

```python
# Plot - Hypthetical 10 year, 5% growth
from matplotlib import rc
import numpy as np
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# rc('font',**{'family':'monospace', 'monospace':['Andale Mono']})
rc('text', usetex=True)

x = np.arange(0, 21, 1, dtype=int)
y = []
y2 = []
g = 0.05
g2 = 0.07

for x_i in x:
    y.append(int(round(1000*(1+g)**x_i, 0)))

for x_i in x:
    y2.append(int(round(1000*(1+g2)**x_i, 0)))

# plot
fig, ax = plt.subplots(figsize=(7,4))

ax.plot(x, y, linewidth=2, color='#00B9FF', label="5\% annual rate")
ax.plot(x, y2, linewidth=2, color='#BA68C8', label="7\% annual rate")

ax.set_xlabel("Year")
ax.set_ylabel("Value \$")
# fig.suptitle("Hypothetical \$1000 Investment with Compound Growth")
ax.set_xticks(np.arange(len(x)))

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.legend(loc=0, frameon=False)


plt.show()
```
